```python
import numpy as np

class C:
    def __init__(self):
        self.x = np.random.randn(5)

class K:
    def __init__(self):
        self.v = None
    
    def eee(self, x, lr = 0.01, momentum = 10):
        if self.v is None:
            self.v = np.zeros_like(x)
        
        self.v = momentum*self.v - lr*x
        x += self.v 

class G:
    def __init__(self):
        self.layer = C()
        self.op = K()

    def momentum(self):
        self.op.eee(self.layer.x)

a = G()
print(a.layer.x)
a.momentum()
print(a.layer.x)
```
