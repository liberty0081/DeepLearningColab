```python
from google.colab import drive
drive.mount('/content/drive')

import sys
sys.path.append('/content/drive/My Drive/Colab Notebooks/NeuralNetwork')

import numpy as np
from keras.datasets import mnist
import keras
from NeuralNetwork import NeuralNetwork

(x_train, t_train), (x_test, t_test) = mnist.load_data()

x_train  = x_train.reshape(60000, 784)
x_test   = x_test.reshape(10000, 784)
x_train  = x_train.astype('float32')
x_test   = x_test.astype('float32')
x_train /= 255
x_test  /= 255
t_train  = keras.utils.np_utils.to_categorical(t_train, 10)
t_test   = keras.utils.np_utils.to_categorical(t_test, 10)

batch_size = 100

net = NeuralNetwork()

net.add_affine(input = x_train.shape[1], output = 50, optimizer="SDG", lr = 0.01)
net.add_batch_norm(size = 50, optimizer = "SDG", lr = 0.01)
net.add_activation_function(func = "ReLU", size = 50)

net.add_affine(input = 50, output = t_train.shape[1], optimizer="SDG", lr = 0.01)
net.add_output_layer(func = "Softmax2CrossEntropy", size = t_train.shape[1])

net.create()
print(str(net.accuracy(x_test, t_test)*100) + "%")  

for epoch in range(10):
    for i in range(100):
        batch_mask = np.random.choice(x_train.shape[0], batch_size)
        x_batch = x_train[batch_mask]
        t_batch = t_train[batch_mask]
        
        net.forward(x_batch, t_batch)
        net.backward()
    
    accuracy = 0

    for j in range(int(x_test.shape[0]/batch_size)):
        test_mask = np.arange(0 + 100*j, 100 + 100*j, 1)
        x_test_batch = x_test[test_mask]
        t_test_batch = t_test[test_mask]

        accuracy += net.accuracy(x_test_batch, t_test_batch)

    print(str(accuracy/(x_test.shape[0]/batch_size)*100) + "%")    
```
